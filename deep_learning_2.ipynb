```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
def subplots(inp, labels):
  for i in range(10):
    x_train_d = inp[labels == i,:,:]
    x_train_i = x_train_d[100,:,:]
    
    plt.figure()
    plt.subplot(2,5,i+1)
    plt.imshow(x_train_i, cmap=plt.get_cmap('gray'))
    plt.title('Label ' + str(i))
    plt.show()

subplots(x_train, y_train)
```

```python
num_train_img = x_train.shape[0]
print(num_train_img)
print(x_train[0:int(0.8 * num_train_img),:,:].shape)
```

```python
num_train_img= x_train.shape[0]
train_ind = np.arange(0, num_train_img)

train_ind_s = np.random.permutation(train_ind)
print(train_ind[1:10])
print(train_ind_s[1:10])

x_train = x_train[train_ind_s,:,:]
y_train = y_train[train_ind_s]

x_valid = x_train[0:int(0.2*num_train_img),:,:]
y_valid = y_train[0:int(0.2*num_train_img)]

x_train = x_train[int(0.2*num_train_img):,:,:]
y_train = y_train[int(0.2*num_train_img):]

print('images in training set: ', x_train.shape[0])
print('images in validation set: ', x_valid.shape[0])
print('images in test set: ', x_test.shape[0])
```

```python
for j in range(3):
  if j == 0:
    print("These are from training data: ")
    subplots(x_train, y_train)

  if j == 1:
    print("These are from validation data: ")
    subplots(x_valid, y_valid)

  if j == 2:
    print("These are from test data: ")
    subplots(x_test, y_test)
```
